```python
"""Scenario 2: Student Marks Performance Analysis
A school wants to analyze exam performance.
Tasks:
Create marks for 100 students across 5 subjects using NumPy.


Store data in Pandas with student IDs.


Calculate:


Total marks


Percentage


Pass/fail (pass ≥ 40%)


Plot:


Histogram of percentages


Bar chart of average marks per subject
"""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#creating data for 100 students for 5 subjects
marks=np.random.randint(0,101,size=(100,5))
marks
```

```python
#creating data 100 students for 5 subjects
df=pd.DataFrame(marks)
df
```

```python
#average marks per subject
avg_marks_per_subject=df.mean(axis=0)
avg_marks_per_subject
```

```python
#calculating total marks for each students
df['total_marks']=df.sum(axis=1)
df
```

```python
#calculating percentage
df['Percentage']=(df['total_marks']/500)*100
df
```

```python
#pass/fail if percentage greater than or equal to 40.
df['Result'] = np.where(df['Percentage'] >= 40, 'Pass', 'Fail')
df
```

```python
#plotting Histogram of Student Percentages
plt.figure(figsize=(8, 5))
plt.hist(df['Percentage'], bins=10, color='skyblue')
plt.xlabel('Percentage')
plt.ylabel('Number of Students')
plt.title('Histogram of Student Percentages')
plt.show()
```

```python
#Average marks per subject
avg_marks_per_subject=df[[0,1,2,3,4]].mean().plot(kind='bar')
plt.title('Average Marks')
plt.show()
```
