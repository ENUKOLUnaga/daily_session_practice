```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data={
    'name':['stephen','robert','joseph','nile','robin'],
    'age':[23,25,26,27,np.nan]
}
data
```

```python
#creating DataFrame
df=pd.DataFrame(data)
df
```

```python
#replace nan values with mean in age
df['age'] = df['age'].fillna(df['age'].mean())
df
```

bar plot - name vs age

```python
#bar plot - name vs age
plt.bar(df['name'],df['age'],color='blue')
plt.title('name vs age')
plt.xlabel('Name')
plt.ylabel('Age')
plt.grid()
plt.show()
```
