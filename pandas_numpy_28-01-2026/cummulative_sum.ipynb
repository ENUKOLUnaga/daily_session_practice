```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "day": ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"],
    "daily_sales": [120, 150, 100, 180, 200, 170, 220]
}
data
```

```python
#creating dataframe
df=pd.DataFrame(data)
df
```

```python
#cummulative sum of daily sales
df["cumulative_sales"] = df["daily_sales"].cumsum()
df
```

```python
#Cumulative Sales Over Time
plt.plot(df["day"], df["cumulative_sales"], marker="o")
plt.xlabel("Day")
plt.ylabel("Cumulative Sales")
plt.title("Cumulative Sales Over Time")
plt.show()
```
