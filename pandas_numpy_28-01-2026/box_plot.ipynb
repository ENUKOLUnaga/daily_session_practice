```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    "class": ["A", "A", "A", "B", "B", "B", "C", "C", "C"],
    "marks": [78, 85, 90, 65, 70, 75, 88, 92, 95]
}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.boxplot(column="marks", by="class", grid=False)

plt.title("Mark Distribution Across Classes")
plt.suptitle("") 
plt.xlabel("Class")
plt.ylabel("Marks")

plt.show()
```
