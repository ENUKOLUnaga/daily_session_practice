```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    'department':['Hr','It','Finance','Hr','It','Finance','Hr','Finance','Hr','It'],
    'salary':[25000,22000,23000,25000,26000,28000,29000,27000,27000,21000]
}
data
```

```python
#creating DataFrame
df=pd.DataFrame(data)
df
```

```python
#Getting average salary from department
avg_salary=df.groupby(df['department'])['salary'].mean()
avg_salary
```

```python
#salary per department
plt.bar(df['department'],df['salary'],color='blue',)
plt.title("salary per department")
plt.xlabel('Department')
plt.ylabel('Salary')
plt.show()
```
