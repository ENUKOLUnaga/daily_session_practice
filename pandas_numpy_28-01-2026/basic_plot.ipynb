Create a DataFrame with columns:
day → 1 to 7


sales → random integers between 100 and 500


Plot a line chart showing sales trend across days.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sale=np.random.randint(100,500,size=(1,7))
sale
```

```python
data={
    "day":[1,2,3,4,5,6,7],
    "sales":np.random.randint(100,501,size=7)
}
data
```

```python
df=pd.DataFrame(data)
df
```

```python
plt.plot(df['day'],df['sales'],marker='o',color='red')
plt.title("sales trends across days")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```
