```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dates = pd.date_range(end=pd.Timestamp.today(), periods=10)
df = pd.DataFrame({
    "date": dates,
    "temperature": [22, 23, 21, 24, 25, 26, 27, 26, 28, 29]
})
df
```

```python
#creating dataframe
df["date"] = pd.to_datetime(df["date"])
df.dtypes
```

```python
#line plot-Temperature Over Last 10 Days
plt.figure(figsize=(8, 4))
plt.plot(df["date"], df["temperature"], marker="o")
plt.title("Temperature Over Last 10 Days")
plt.xlabel("Date")
plt.ylabel("Temperature in celsius")
plt.xticks(rotation=90)
plt.grid(True)
```
