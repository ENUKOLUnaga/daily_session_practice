```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= {
    "city": ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"],
    "population": [8419600, 3980400, 2716000, 2328000, 1690000]
}
data
```

```python
#creating dataframe
df=pd.DataFrame(data)
df
```

```python
#filtering data who are greater than 1 million
filtered_df = df[df["population"] > 1_000_000]
filtered_df
```

```python
#barchart - Cities with Population Greater Than 1 Million
plt.bar(filtered_df["city"], filtered_df["population"], color="orange")
plt.xlabel("City")
plt.ylabel("Population")
plt.title("Cities with Population Greater Than 1 Million")
plt.xticks(rotation=45)
```
